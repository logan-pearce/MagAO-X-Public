```python
import pickle
pickles = pickle.load(open('pickle_models.pkl','rb'))
spt = [key for key in pickles.keys()]
```

```python
import pandas as pd
# Get MagAOX filter curves:
directory = ''
#magaox_sci2-gp_bs-65-35_scibs-5050.dat
file = 'magaox_sci1-zp_bs-65-35_scibs-5050.dat'
zfilter = pd.read_table(directory+file, comment='#', 
                  names=['wavelength [m]','transmission'], delim_whitespace=True)
zfilter['normalized transmission'] = zfilter['transmission']/np.max(zfilter['transmission'])

file = 'magaox_sci2-ip_bs-65-35_scibs-5050.dat'
ifilter = pd.read_table(directory+file, comment='#', 
                  names=['wavelength [m]','transmission'], delim_whitespace=True)
ifilter['normalized transmission'] = ifilter['transmission']/np.max(ifilter['transmission'])

file = 'magaox_sci1-rp_bs-65-35_scibs-5050.dat'
rfilter = pd.read_table(directory+file, comment='#', 
                  names=['wavelength [m]','transmission'], delim_whitespace=True)
rfilter['normalized transmission'] = rfilter['transmission']/np.max(rfilter['transmission'])

file = 'magaox_sci2-gp_bs-65-35_scibs-5050.dat'
gfilter = pd.read_table(directory+file, comment='#', 
                  names=['wavelength [m]','transmission'], delim_whitespace=True)
gfilter['normalized transmission'] = gfilter['transmission']/np.max(gfilter['transmission'])


# Get Sloan filter curves:
directory = '/Users/loganpearce/Dropbox/Uarizona/research/filter_curves/'

file = 'SDSS_z.dat'
SDSSzfilter = pd.read_table(directory+file, comment='#', 
                  names=['wavelength [um]','transmission'], delim_whitespace=True)
SDSSzfilter['normalized transmission'] = SDSSzfilter['transmission']/np.max(SDSSzfilter['transmission'])

file = 'SDSS_i.dat'
SDSSifilter = pd.read_table(directory+file, comment='#', 
                  names=['wavelength [um]','transmission'], delim_whitespace=True)
SDSSifilter['normalized transmission'] = SDSSifilter['transmission']/np.max(SDSSifilter['transmission'])

file = 'SDSS_r.dat'
SDSSrfilter = pd.read_table(directory+file, comment='#', 
                  names=['wavelength [um]','transmission'], delim_whitespace=True)
SDSSrfilter['normalized transmission'] = SDSSrfilter['transmission']/np.max(SDSSrfilter['transmission'])

file = 'SDSS_g.dat'
SDSSgfilter = pd.read_table(directory+file, comment='#', 
                  names=['wavelength [um]','transmission'], delim_whitespace=True)
SDSSgfilter['normalized transmission'] = SDSSgfilter['transmission']/np.max(SDSSgfilter['transmission'])

%matplotlib inline
plt.figure(figsize=(9,5))
plt.plot(gfilter['wavelength [m]']*u.m.to(u.um),gfilter['normalized transmission'],label='Mg$^\prime$')
plt.plot(rfilter['wavelength [m]']*u.m.to(u.um),rfilter['normalized transmission'],label='Mr$^\prime$')
plt.plot(ifilter['wavelength [m]']*u.m.to(u.um),ifilter['normalized transmission'],label='Mi$^\prime$')
plt.plot(zfilter['wavelength [m]']*u.m.to(u.um),zfilter['normalized transmission'],label='Mz$^\prime$')

plt.plot(SDSSgfilter['wavelength [um]'],SDSSgfilter['normalized transmission'],label='Sg$^\prime$',ls='--')
plt.plot(SDSSrfilter['wavelength [um]'],SDSSrfilter['normalized transmission'],label='Sr$^\prime$',ls='--')
plt.plot(SDSSifilter['wavelength [um]'],SDSSifilter['normalized transmission'],label='Si$^\prime$',ls='--')
plt.plot(SDSSzfilter['wavelength [um]'],SDSSzfilter['normalized transmission'],label='Sz$^\prime$',ls='--')

plt.xlabel('Wavelength [$\mu$m]')
plt.ylabel('Transmission')
plt.title('MagAO-X + Sloan filter transmission')
plt.legend(loc=(1.01,0))
plt.tight_layout()
plt.grid(ls=":")
#plt.savefig(directory+'MagAOXfilters_SDSSfilters.png',dpi=300)
plt.show()
```

```python
from scipy.interpolate import interp1d
colors_z = np.array([])
colors_i = np.array([])
colors_r = np.array([])
colors_g = np.array([])


for SpT in spt:
    s = pickles[SpT]
    func = interp1d(s['wavelength'], s['flux'])
    ### MagAO X fluxes:
    # interpolate the model onto the grid of wavelengths from the filter transmission:
    s_resampled_onto_Zfilter_wavelengths = func(zfilter['wavelength [m]']*u.m.to(u.AA))
    s_resampled_Zwavelengths = zfilter['wavelength [m]']*u.m.to(u.AA)
    zfilter['normalized transmission'] = zfilter['transmission']/np.max(zfilter['transmission'])

    s_resampled_onto_Ifilter_wavelengths = func(ifilter['wavelength [m]']*u.m.to(u.AA))
    s_resampled_Iwavelengths = ifilter['wavelength [m]']*u.m.to(u.AA)
    ifilter['normalized transmission'] = ifilter['transmission']/np.max(ifilter['transmission'])

    s_resampled_onto_Rfilter_wavelengths = func(rfilter['wavelength [m]']*u.m.to(u.AA))
    s_resampled_Rwavelengths = rfilter['wavelength [m]']*u.m.to(u.AA)
    rfilter['normalized transmission'] = rfilter['transmission']/np.max(rfilter['transmission'])

    s_resampled_onto_Gfilter_wavelengths = func(gfilter['wavelength [m]']*u.m.to(u.AA))
    s_resampled_Gwavelengths = gfilter['wavelength [m]']*u.m.to(u.AA)
    gfilter['normalized transmission'] = gfilter['transmission']/np.max(gfilter['transmission'])

    # Integrate flux in magaox filters:
    dl = np.mean(np.array([s_resampled_Zwavelengths[i] - s_resampled_Zwavelengths[i-1] 
                   for i in range(1,len(s_resampled_Zwavelengths))]))

    MagAOXZmodel_flux = np.sum(s_resampled_onto_Zfilter_wavelengths*zfilter['normalized transmission']*dl)
    MagAOXImodel_flux = np.sum(s_resampled_onto_Ifilter_wavelengths*ifilter['normalized transmission']*dl)
    MagAOXRmodel_flux = np.sum(s_resampled_onto_Rfilter_wavelengths*rfilter['normalized transmission']*dl)
    MagAOXGmodel_flux = np.sum(s_resampled_onto_Gfilter_wavelengths*gfilter['normalized transmission']*dl)

    ## Repeat for Sloan filters:

    # interpolate the model onto the grid of wavelengths from the filter transmission:
    s_resampled_onto_SDSSZfilter_wavelengths = func(SDSSzfilter['wavelength [um]']*u.um.to(u.AA))
    s_resampled_SDSSZwavelengths = SDSSzfilter['wavelength [um]']*u.um.to(u.AA)
    SDSSzfilter['normalized transmission'] = SDSSzfilter['transmission']/np.max(SDSSzfilter['transmission'])

    s_resampled_onto_SDSSIfilter_wavelengths = func(SDSSifilter['wavelength [um]']*u.um.to(u.AA))
    s_resampled_SDSSIwavelengths = SDSSifilter['wavelength [um]']*u.um.to(u.AA)
    SDSSifilter['normalized transmission'] = SDSSifilter['transmission']/np.max(SDSSifilter['transmission'])

    s_resampled_onto_SDSSRfilter_wavelengths = func(SDSSrfilter['wavelength [um]']*u.um.to(u.AA))
    s_resampled_SDSSRwavelengths = SDSSrfilter['wavelength [um]']*u.um.to(u.AA)
    SDSSrfilter['normalized transmission'] = SDSSrfilter['transmission']/np.max(SDSSrfilter['transmission'])

    s_resampled_onto_SDSSGfilter_wavelengths = func(SDSSgfilter['wavelength [um]']*u.um.to(u.AA))
    s_resampled_SDSSGwavelengths = SDSSgfilter['wavelength [um]']*u.um.to(u.AA)
    SDSSgfilter['normalized transmission'] = SDSSgfilter['transmission']/np.max(SDSSgfilter['transmission'])

    # Integrate:
    dl = np.mean(np.array([s_resampled_SDSSZwavelengths[i] - s_resampled_SDSSZwavelengths[i-1] 
                   for i in range(1,len(s_resampled_SDSSZwavelengths))]))

    SDSSZmodel_flux = np.sum(s_resampled_onto_SDSSZfilter_wavelengths*SDSSzfilter['normalized transmission']*dl)
    SDSSImodel_flux = np.sum(s_resampled_onto_SDSSIfilter_wavelengths*SDSSifilter['normalized transmission']*dl)
    SDSSRmodel_flux = np.sum(s_resampled_onto_SDSSRfilter_wavelengths*SDSSrfilter['normalized transmission']*dl)
    SDSSGmodel_flux = np.sum(s_resampled_onto_SDSSGfilter_wavelengths*SDSSgfilter['normalized transmission']*dl)

    MagAOXtoSDSS_Z = -2.5*np.log10(SDSSZmodel_flux) - (-2.5*np.log10(MagAOXZmodel_flux))
    MagAOXtoSDSS_I = -2.5*np.log10(SDSSImodel_flux) - (-2.5*np.log10(MagAOXImodel_flux))
    MagAOXtoSDSS_R = -2.5*np.log10(SDSSRmodel_flux) - (-2.5*np.log10(MagAOXRmodel_flux))
    MagAOXtoSDSS_G = -2.5*np.log10(SDSSGmodel_flux) - (-2.5*np.log10(MagAOXGmodel_flux))
    
    colors_g = np.append(colors_g,MagAOXtoSDSS_G)
    colors_r = np.append(colors_r,MagAOXtoSDSS_R)
    colors_i = np.append(colors_i,MagAOXtoSDSS_I)
    colors_z = np.append(colors_z,MagAOXtoSDSS_Z)
    
    
p_wfs = pd.DataFrame()
p_wfs['SpT'],p_wfs['g'],p_wfs['r'],p_wfs['i'],p_wfs['z'] = spt,colors_g,colors_r,colors_i,colors_z
```

```python
dwarfs = [i for i in p_wfs['SpT'] if 'V' in i]
giants = [i for i in p_wfs['SpT'] if 'III' in i]

dwarfs_colors_g = [p_wfs['g'][i] for i in range(len(dwarfs))]
giants_colors_g = [p_wfs['g'][i] for i in range(len(dwarfs),len(p_wfs))]

dwarfs_colors_r = [p_wfs['r'][i] for i in range(len(dwarfs))]
giants_colors_r = [p_wfs['r'][i] for i in range(len(dwarfs),len(p_wfs))]

dwarfs_colors_i = [p_wfs['i'][i] for i in range(len(dwarfs))]
giants_colors_i = [p_wfs['i'][i] for i in range(len(dwarfs),len(p_wfs))]

dwarfs_colors_z = [p_wfs['z'][i] for i in range(len(dwarfs))]
giants_colors_z = [p_wfs['z'][i] for i in range(len(dwarfs),len(p_wfs))]
```

```python
spt_letter_conv = {'O':0,'B':1,'A':2,'F':3,'G':4,'K':5,'M':6}

spt_numbers = np.array([])
for s in spt:
    letter = s[0]
    number = spt_letter_conv[letter]
    
    type_number = np.float(s[1]) / 10
    
    spt_numbers = np.append(spt_numbers,number + type_number)
    
spt_numbers[-1] = 7.0
spt_numbers[5] = 1.6
spt_numbers_dwarfs = spt_numbers[range(len(dwarfs))]
spt_numbers_giants = spt_numbers[range(len(dwarfs),len(spt_numbers))]
spt_numbers_dwarfs,spt_numbers_giants
```

```python
p_wfs['SpT Number'] = spt_numbers
directory = ''
p_wfs.to_csv(directory+'MagAOXfilters_toSDSSfilters_color_conversion.csv',index=False)
```
